```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
    SELECT name, SUM( revenue ) AS totalrevenue
FROM (

SELECT name, (

CASE WHEN B.memid =0
THEN (
guestcost * slots
)
ELSE (
membercost * slots
)
END
) AS revenue
FROM Facilities AS F
JOIN Bookings AS B ON B.facid = F.facid
) AS sub
GROUP BY name
HAVING totalrevenue <1000
        """
    query2 = """
       SELECT CONCAT( R.firstname, ' ', R.surname ) AS membername, CONCAT( M.firstname, ' ', M.surname ) AS referedby
FROM Members AS M
JOIN Members AS R ON R.recommendedby = M.memid
WHERE R.recommendedby !=0
ORDER BY R.surname, R.firstname
        """
    query3 = """
        SELECT DISTINCT F.facid,name, firstname, surname
FROM Members AS M
JOIN Bookings AS B ON M.memid = B.memid
JOIN Facilities AS F ON F.facid = B.facid
WHERE M.memid !=0
        """
    query4 = """
SELECT F.facid, name, SUM( slots ) , EXTRACT(MONTH FROM starttime ) AS mth FROM Bookings AS B
JOIN Facilities AS F ON B.facid = F.facid
GROUP BY mth, F.facid"""
    
    cur.execute(query4)
   

 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
